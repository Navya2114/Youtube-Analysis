```python
filtered=data[['views','dislikes','comment_count','Genre','likes']]
filtered
```

```python
f1=pd.get_dummies(filtered['Genre'])
f1
```

```python
f1.drop('Travel & Events',axis=1,inplace=True)
f1
```

```python
fi=pd.concat([filtered,f1],axis=1)
fi
```

```python
fi.drop('Genre',axis=1,inplace=True)
fi
```

```python
fi.to_csv('filtered.csv',index=False)
```

```python
x=fi.drop('likes',axis=1)
x
```

```python
y=fi['likes']
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr=LinearRegression()
```

```python
lr.fit(x_train,y_train)
```

```python
lr.predict(x_test.head())
```

```python
np.array(y_test.head())
```

```python
lr.score(x_test,y_test)
```
